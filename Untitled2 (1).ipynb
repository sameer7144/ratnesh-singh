```python
class Tree:
    def __init__(self, name, height_feet, fruit):
        self.name = name
        self.height_feet = height_feet
        self.fruit = fruit

    def __str__(self):
        return f"Tree Name: {self.name}, Height: {self.height_feet} ft, Fruit: {self.fruit}"


class TreeDatabase:
    def __init__(self):
        self.trees = []

    def add_tree(self, name, height_feet, fruit):
        tree = Tree(name, height_feet, fruit)
        self.trees.append(tree)
        print(f"\n‚úî Tree '{name}' added successfully!\n")

    def show_all_trees(self):
        if not self.trees:
            print("\nNo trees available.\n")
            return
        
        print("\nüå≥ List of All Trees:\n")
        for t in self.trees:
            print(t)

    def search_tree(self, name):
        for t in self.trees:
            if t.name.lower() == name.lower():
                print("\nüîç Tree Found:\n")
                print(t)
                return
        print("\n‚ùå Tree not found.\n")


# Main Program
db = TreeDatabase()

while True:
    print("""
===== TREE INFORMATION SYSTEM =====
1. Add New Tree
2. Show All Trees
3. Search Tree
4. Exit
""")

    choice = input("Enter your choice: ")

    if choice == "1":
        name = input("Enter tree name: ")
        height = float(input("Enter tree height (in feet): "))
        fruit = input("Enter tree fruit: ")
        db.add_tree(name, height, fruit)

    elif choice == "2":
        db.show_all_trees()

    elif choice == "3":
        search = input("Enter tree name to search: ")
        db.search_tree(search)

    elif choice == "4":
        print("Exiting program... Goodbye!")
        break

    else:
        print("Invalid choice. Please try again.")
```
